```python
%matplotlib notebook
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython import display
```

```python
def interact_scatter_line(a, b, ax):
    ax.clear()
    min_x, max_x = -4.5, 11.5
    rng          = np.random.RandomState(1)
    x            = (max_x - min_x) * rng.rand(100) + min_x
    y_scatter    = 1/3 * x ** 2 - 3 * x - 7 + 6 * (rng.rand(100) - 0.5)
    min_y, max_y = y_scatter.min(), y_scatter.max()
    y_line       = a * x + b
    plt.xlim([min_x, max_x])
    plt.ylim([min_y - 0.5, max_y + 0.5])
    ax.scatter(x, y_scatter)
    ax.plot(x, y_line)
```

```python
fig, ax = plt.subplots()
interact(interact_scatter_line, a = (-15, 15, 1), b = (-30, 30, 1), ax = fixed(ax))
```

```python
def interact_scatter_poly(a, b, c, ax):
    ax.clear()
    min_x, max_x = -4.5, 11.5
    rng          = np.random.RandomState(1)
    x            = (max_x - min_x) * rng.rand(100) + min_x
    y_scatter    = 0.3 * x ** 2 - 3 * x - 7 + 6 * (rng.rand(100) - 0.5)
    min_y, max_y = y_scatter.min(), y_scatter.max()
    x_poly       = np.linspace(min_x, max_x, 200) 
    y_poly       = a * x_poly ** 2 + b * x_poly + c
    plt.xlim([min_x, max_x])
    plt.ylim([min_y - 0.5, max_y + 0.5])
    ax.scatter(x, y_scatter)
    ax.plot(x_poly, y_poly)
```

```python
fig, ax = plt.subplots()
interact(interact_scatter_poly, a = (-2, 2, 0.02), b = (-15, 15, 0.2), c = (-30, 30, 0.2), ax = fixed(ax))
```
